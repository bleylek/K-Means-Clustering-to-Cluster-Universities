# K Means Cluster Creation

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df.drop('Private', axis=1))
```

```python
# Cluster center vectors
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

# Evaluation

There is no perfect way to evaluate clustering if we don't have the labels, however since this is just an exercise, we do have the labels, so we take advantage of this to evaluate our clusters. However we usually won't have this luxury in the real world.

Creating a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.

```python
def converter(private):
  if private == 'Yes':
    return 1
  else:
    return 0
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
df.head()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(df['Cluster'],kmeans.labels_))
```

```python
print(classification_report(df['Cluster'],kmeans.labels_))
```
